## Background

This problem is based on Exercise 8.2 in Newman's *Computational Physics*. The **Lotka-Volterra** equations are a model of the population dynamics of predators and prey. See: https://en.wikipedia.org/wiki/Lotka–Volterra_equations

Let $x$ be the population of [Caterpie](http://www.pokemon.com/us/pokedex/caterpie) (prey).  
Let $y$ be the population of [Pidgeotto](http://www.pokemon.com/us/pokedex/pidgeotto) (predator).

In this system, the two main factors that govern the caterpie population is their reproduction rate and the rate at which they are [eaten by pidgeottos](https://bulbapedia.bulbagarden.net/wiki/Pokémon_food). Analogously, the pidgeotto population is governed by the availability of food and the desire to move elsewhere if there is too much competition for food. 

The Lotka-Volterra equations are:  

$\displaystyle \frac{dx}{dt} = a x - b x y$  

$\displaystyle \frac{dy}{dt} = c x y - d y$  
 
The dimensionless parameters $a$, $b$, $c$, and $d$ are positive real numbers that characterize the interaction of the two species.

## 1. Understanding the problem

* $a$ is the rate at which caterpie reproduce. This is because the rate of change of the caterpie population ($\dot{x}$) has a term $ax$; this term makes the population increase at a rate proportional to the population itself. 
* $b$ is the rate of predation of caterpie by pigetto. This is because the caterpie population would decrease , ($\dot{x} < 0$), if $b$ is a positive number. If $x$ or $y = 0$, there can be no predation. Therefore, the predation of caterpie is proportional to $bxy$. 
* $c$ is the growth rate of the predator, pigetto. We can see that the growth of pigetto depends on the availability of caterpie for them to prey on. If there are pigetto, that is if $x \geq 0$, the pigetto population will continue to grow. 
* $d$ is the loss rate of caterpie for the pigetto to prey on. Caterpie could either be migrating or be dying out because of natural causes. This leads to an exponential decay in the pigetto population. 
* The special case $b = c = 0$ is analytically solvable. In that case the populations don't affect each other. **FILL THIS IN:** *Explain what happens to each population* In this case, the caterpie population would increase exponentially, $x(t) = \exp(at)$. The caterpie population increases because the pigetto are not eating them. On the other hand, the pigetto would slowly die out. They are not able to prey upon the caterpie. Their population would decrease like this: $y(t) = \exp(-dt)$

## 2. Code it up!

Let's build the code to understand this system. We have two first order differential equations:

$$\displaystyle \frac{dx}{dt} = f(x,y,t)$$

$$\displaystyle \frac{dy}{dt} = g(x,y,t)$$

Let these be functions of global variables, `a`, `b`, `c`, and `d`. Fill in the sub-problems below.

### 2.a. Parameters, functions

````python
# a quick way to fix these parameters in one line
a, b, c, d = .6, 1.3, 1.0, 1.0

def f(x,y,t):
    '''
    This is the rate of change of Caterpie as a function 
    of the current number of Caterpie (x), the current number
    of Pigeotto (y), and the current time (t).
    '''
    return a*x - b*x*y

def g(x,y,t):
    '''
    This is the rate of change of Pigeotto as a function 
    of the current number of Caterpie (x), the current number
    of Pigeotto (y), and the current time (t).
    '''
    return c*x*y - d*y
````

````python
# Test your code: you should get 0.047 and -0.09:
print( f(.1,.1,0) )
print( g(.1,.1,0) )
````

### 2.b. Time evolution 

At $t=0$, use the **initial conditions** $x_0=y_0=2$. We would like to know what happens to the populations after some amount of. We won't be picky about units, so let's assume that $t$ implicitly has some agreed upon units of time. (Thus $a$, $b$, $c$, and $d$ all have units of rate.)

1. Initialize a Python array with 100 elements. I suggest using `np.arange`. 

2. Use **Euler's method** ("simplest, most obvious ODE algorithm") to fill Python arrays `xlist` and `ylist`. Use $\Delta t = 0.05$.

3. Plot the populations as a function of time using `matplotlib`. 

````python
import numpy as np

a, b, c, d = .6, 1.3, 1.0, 1.0

x0 = 2
y0 = 2
dt = .05
N  = 100

## A an array: [0,1,2, ..., N]
time = np.linspace(0,100,N)


xlist = []
ylist = []

# Initialize x and y
x = x0
y = y0

for t in time:
    xlist.append(x)
    ylist.append(y)
    x += f(x,y,t)*dt
    y += g(x,y,t)*dt


%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(time, xlist, label="Caterpie")
plt.plot(time, ylist, label="Pidgeotto")
plt.legend()
plt.show()
````

### 2.c. Discussion

At first, there are a lot of Pidgeotto and they all prey on the poor Caterpie. Therefore, the population of the Pidgeotto increases (they have food), and the population of the Caterpie decreases (they are food). However, the population of the Caterpie gets dangerously low and the (huge) population of Pidgeotto have a difficult time finding food. (Take that!). So the Pidgeotto population decreases because many of them die of hunger. After the Pidgeotto population has decreased in number, the Caterpie population stops dying out and then starts increasing. Then at one point the Caterpie population overtakes that of the Pidgeotto. In the future, the Caterpie population would increase greatly, but then again start to be eaten by the dwindling number of Pidgeotto. Then the two populations would trade places and the cycle continues...

### 2.f. [Things Fall Apart](https://en.wikipedia.org/wiki/Things_Fall_Apart)

Let's see what happens if we choose our integration poorly. 

Original set of  Lotka-Volterra parameters:

```python
a, b, c, d = .6, 1.3, 1, 1
```

But this time, let's try to be cheap and use a much bigger time step, using `dt = 0.5` and `N = 10`. This should look nothing like the behavior we expect from the nice plots we produced earlier.

````python
import numpy as np

# A reminder of our earlier parameters
a, b, c, d = .6, 1.3, 1.0, 1.0

x0 = 2
y0 = 2
dt = .5
N  = 10

## A an array: [0,1,2, ..., N]
time = np.linspace(0,100,N)


xlist = []
ylist = []

# Initialize x and y
x = x0
y = y0

for t in time:
    xlist.append(x)
    ylist.append(y)
    x += f(x,y,t)*dt
    y += g(x,y,t)*dt


%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(time, xlist, label="Caterpie")
plt.plot(time, ylist, label="Pidgeotto")
plt.legend()
plt.show()
````

**What went wrong**: The $dt$ term is simply too large. After the first iteration of the Euler method, when $t = 11.111...$, the caterpie population goes to 0. 2 caterpie are eaten in this time frame (the whole population). Once the population goes extinct, there is no way there can be anymore Caterpie anywhere (*conservation of Caterpie*). The Pidgeotto need to feed on Caterpie for food, or else they will die out. So they go extinct soon after. 

This is all because the $dt$ is too large. If we make that smaller, and increase the number of samples, ($N \geq 10$), we will get a smoother graph where no Pokemon go extinct!

## 3. Population Dynamics Phase Space

Notice the periodic behavior of the system. The plots that we're making are useful if we really care about what time of year to visit [Kanto](https://bulbapedia.bulbagarden.net/wiki/Region) in order to see caterpies or pigeottos, but are not actually as useful for understanding how the two populations are inter-related. Let's make those plots. 

### 3.a. The caterpie---pigetotto plane

Code from Problem 2.d (with `N=500`). Plots the caterpie and pidgeotto populations on the horizontal and vertical axes, respectively. 

````python
import numpy as np

a, b, c, d = .6, 1.3, 1.0, 1.0

x0 = 2
y0 = 2
dt = .05
N  = 500  #CHANGED ###DIFFERENT

## A an array: [0,1,2, ..., N]
time = np.linspace(0,100,N)


xlist = []
ylist = []

# Initialize x and y
x = x0
y = y0

for t in time:
    xlist.append(x)
    ylist.append(y)
    x += (a*x - b*y*x)*dt
    y += (c*x*y - d*y)*dt


%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(xlist,ylist)
plt.title("Pidgeotto v Caterpie Population")
plt.xlabel("Caterpie")
plt.ylabel("Pidgeotto")
plt.show()
````

### 3.b. The velocity in population space

**Quiver plots** are what `matplotlib` calls plots of vector fields. Here's an example of how they work:

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# base positions of the vectors
x = np.linspace(0,1,5)
y = np.linspace(0,1,5)

# magnitude of the x- and y-components of each vector
u = np.zeros((5,5)) # initialize to zero everywhere
v = np.zeros((5,5)) # initialize to zero everywhere

u[2,2] = .1 # x-component of vector at (2,2)
v[2,2] = .2 # y-component of vector at (2,2)

u[2,3] = .2 # x-component of vector at (2,2)
v[2,3] = .1 # y-component of vector at (2,2)

plt.quiver(x, y, u, v, scale=1)
plt.show()
````

Vector field that shows the vector 

$$ \mathbf{v}(x,y) = \left (\dot x, \dot y \right)$$

at each position on the caterpie-pigeotto plane.

````python
ff = np.vectorize(f)
gg = np.vectorize(g)

side = np.linspace(0,4,10)
XX,YY = np.meshgrid(side,side/2)
UU = ff(XX,YY,0)
VV = gg(XX,YY,0)

plt.quiver(XX,YY,UU,VV)

plt.plot(xlist,ylist)
plt.title("Pidgeotto v Caterpie Population")
plt.xlabel('Caterpie')
plt.ylabel('Pidgeotto')
plt.show()
````

This image now shows the cyclicity of the caterpie and pigeotto populations as well as how long the populations spend in any particular part of the cycle.

### 3.c. Describe the dynamics

**Phase 1:** Phase 1 occurs in the upper Right Hand Corner of the plot. This is when both populations are thriving. This phase begins with a surplus of Caterpie. They are quickly eaten by Pidgeotto, and the two populations spend a little bit of time with equal numbers. But very quickly, the growing number of Pidgeotto eat and eat Caterpie until there are not very many Caterpie left.

Summary: Fast transition from surplus of caterpie to surplus of Pidgeotto. 

**Phase 2:** Phase 2 begins when the rate of change of Pidgeotto population decreases rapidly in the left hand side of the graph. The Pidgeotto population grew very large and ate almost all the Caterpie. There is not alot of food left for the Pidgeotto, so they begin to die out. This continues up until the two populations are basically equal in number.

summary: The Pidgeotto population slowly dies out because of a low number of Caterpie to prey on.

**Phase 3:** Phase 3 begins when the two populations are basically equal in number, and are very small. This can be seen in the lower left corner of the graph. The caterpie population slowly increases because there are not many Pidgeotto there to eat them. 

summary: Caterpie population slowly regenerates due to low number of predators eating them.

## 4. External Effects

Modify $f(x,y,t)$ and $g(x,y,t)$ to include external effects.

Changing $f(x,y,t)$ to include seasonal migration of caterpie (given by $sin(\frac{t}{25 \pi})$).

````python
def f(x,y,t):
    '''
    This is the rate of change of Caterpie as a function 
    of the current number of Caterpie (x), the current number
    of Pigeotto (y), and the current time (t).
    '''
    return a*x - b*x*y*np.sin(t/25*np.pi)

def g(x,y,t):
    '''
    This is the rate of change of Pigeotto as a function 
    of the current number of Caterpie (x), the current number
    of Pigeotto (y), and the current time (t).
    '''
    return c*x*y - d*y
````

````python
import numpy as np

a, b, c, d = .6, 1.3, 1.0, 1.0

x0 = 2
y0 = 2
dt = .05
N  = 100

## A an array: [0,1,2, ..., N]
time = np.linspace(0,100,N)


xlist = []
ylist = []

# Initialize x and y
x = x0
y = y0

for t in time:
    xlist.append(x)
    ylist.append(y)
    x += f(x,y,t)*dt
    y += g(x,y,t)*dt


%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(time, xlist, label="Caterpie")
plt.plot(time, ylist, label="Pidgeotto")
plt.legend()
plt.show()
````
